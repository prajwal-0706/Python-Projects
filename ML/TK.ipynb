```python
import tkinter as tk
import os
```

```python
def on_button_click_1():
    result_label.config(text="Opening the Notepad")
    os.system("notepad")
    
def on_button_click_2():
    result_label.config(text="Opening the Notepad")
    os.system("start iexplore")
    
def on_button_click_3():
    result_label.config(text="Opening the Calculator")
    os.system("calc")
    
def on_button_click_4():
    result_label.config(text="Opening the VS Code")
    os.system("code .")
        
def main():
    global root, result_label

    root = tk.Tk()
    root.title("Menu of Team 10")
    root.minsize(700,700)

    label = tk.Label(root, text="Menu", font= ("Arial", 20))
    label.pack()
    
    button_style_1 = {
        "font": ("Arial", 12),
        "bg": "#4CAF50",  
        "fg": "white",    
        "bd": 2,         
    }
    
    button_style_2 = {
        "font": ("Arial", 12),
        "bg": "#219EBC",  
        "fg": "white",    
        "bd": 2,         
    }
    
    button_style_3 = {
        "font": ("Arial", 12),
        "bg": "#fb8500",  
        "fg": "white",    
        "bd": 2,         
    }
    
    button_style_4 = {
        "font": ("Arial", 12),
        "bg": "#eaac8b",  
        "fg": "white",    
        "bd": 2,         
    }
    

    button = tk.Button(root, text="Open the Notepad", command=on_button_click_1,**button_style_1)
    button.pack(pady=10)
    
    button1 = tk.Button(root, text="Open the explorer", command=on_button_click_2, **button_style_2)
    button1.pack(pady=10)
    
    button2 = tk.Button(root, text="Open the Calculator", command=on_button_click_3, **button_style_3)
    button2.pack(pady=10)
    
    button3 = tk.Button(root, text="Open the VS Code", command=on_button_click_4, **button_style_4)
    button3.pack(pady=10)

    result_label = tk.Label(root, text="", fg="red")
    result_label.pack()

    root.mainloop()


main()
```

```python
import tkinter as tk
from tkinter import ttk


root = tk.Tk()
root.title("Menu Design")


background_image = tk.PhotoImage(file="image.jpg", master = root)  # Replace with your image file
background_label = tk.Label(root, image=background_image)
background_label.place(relwidth=1, relheight=1)

# Create a frame for buttons
button_frame = ttk.Frame(root)
button_frame.place(relx=0.5, rely=0.5, anchor="center")

# Define button functions
def button1_action():
    print("Button 1 clicked")

def button2_action():
    print("Button 2 clicked")

def button3_action():
    print("Button 3 clicked")

# Create buttons with styles
button1 = ttk.Button(button_frame, text="Button 1", command=button1_action)
button2 = ttk.Button(button_frame, text="Button 2", command=button2_action)
button3 = ttk.Button(button_frame, text="Button 3", command=button3_action)

button1.pack(pady=10)
button2.pack(pady=10)
button3.pack(pady=10)

# Start the GUI event loop
root.mainloop()
```
